```python

#Importing the packages

from bs4 import BeautifulSoup
import requests
import pandas as pd
import urllib.parse
```

HTTP Request

```python
# Store the website in a variable

website = 'https://www.laptopsdirect.co.uk/ct/laptops-and-netbooks/laptops'
headers = requests.utils.default_headers()
headers.update({
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/105.0.0.0 Safari/537.36',
    'Accept-Language': 'en-GB,en-US;q=0.9,en;q=0.8',
    'Referer': 'https://www.google.com/',
    'DNT': '1',
    'Pius' : 'subtomeplease',
})
```

```python
#Get request

response = requests.get(website, headers = headers)
```

```python
# Status code

response.status_code
```

```python
# Soup object

soup = BeautifulSoup(response.content, 'html.parser')
soup
```

Results

```python
results = soup.find_all('div', attrs = {'class' : 'OfferBox'})
len(results)
```

```python
results[0]
```

Getting the necessary data

```python
# Name of product

results[0].find('a', attrs = {'class' : 'offerboxtitle'}).text
```

```python
# Price of product

results[0].find('span', attrs = {'class' : 'offerprice'}).text.strip()
```

```python
# Rating of product

results[0].find('star-rating').get('rating-value')
```

```python
# Review count

results[0].find('star-rating').get('ratings-count')
```

```python
# Relative URL

Relative_url = results[0].find('a', attrs = {'class' : 'offerboxtitle'}).get('href')
```

```python
# Root URL

Root_url = 'https://www.laptopsdirect.co.uk'
```

```python
# Combine root URL and relative URL

Complete_URL = Root_url + Relative_url
Complete_URL
```

```python
# Product description

results[0].find('div', attrs = {'class' : 'productInfo'}).text.strip().replace('\n', ', ')
```

Using a For Loop to Scrape the Data

```python
product_name = []
product_price = []
review_rating = []
review_count = []
relative_url = []
product_description = []


for result in results:
    
    # Name of product
    try:
        product_name.append(result.find('a', attrs = {'class' : 'offerboxtitle'}).text)
    except:
        product_name.append('N/A')
    
    # Price of product
    try:
        product_price.append(result.find('span', attrs = {'class' : 'offerprice'}).text.strip())
    except:
        product_price.append('N/A')
        
    # Rating of product
    try:
        review_rating.append(result.find('star-rating').get('rating-value'))
    except:
        review_rating.append('N/A')
    
    # Review count
    try:
        review_count.append(result.find('star-rating').get('ratings-count'))
    except:
        review_count.append('N/A')
    
    # Relative URL
    try:
        relative_url.append(result.find('a', attrs = {'class' : 'offerboxtitle'}).get('href'))
    except:
        relative_url.append('N/A')
    
    # Product details
    try:
        product_description.append(result.find('div', attrs = {'class' : 'productInfo'}).text.strip().replace('\n', ', '))
    except:
        product_description.append('N/A')
```

```python
# Combine URLs

url_combined = []

for link in relative_url:
    url_combined.append(Root_url + link)

url_combined
```

DataFrame 

```python
products_overview = pd.DataFrame({'Product Name' : product_name, 'Product Price' : product_price, 'Review Rating' : review_rating, 'Review Count' : review_count, 'Product Description' : product_description, 'Product URL' : url_combined})
products_overview
```

Scraping Many Pages

```python
product_name = []
product_price = []
review_rating = []
review_count = []
relative_url = []
product_description = []

for i in range(1,49):
    
    # Website
    website = 'https://www.laptopsdirect.co.uk/ct/laptops-and-netbooks/laptops?page=' + str(i)
     
    # Get request
    response = requests.get(website, headers = headers)
    
    # Soup object
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Results
    results = soup.find_all('div', attrs = {'class' : 'OfferBox'})
    
    # Loop through results
    for result in results:
    
        # Name of product
        try:
            product_name.append(result.find('a', attrs = {'class' : 'offerboxtitle'}).text)
        except:
            product_name.append('N/A')

        # Price of product
        try:
            product_price.append(result.find('span', attrs = {'class' : 'offerprice'}).text.strip())
        except:
            product_price.append('N/A')

        # Rating of product
        try:
            review_rating.append(result.find('star-rating').get('rating-value'))
        except:
            review_rating.append('N/A')

        # Review count
        try:
            review_count.append(result.find('star-rating').get('ratings-count'))
        except:
            review_count.append('N/A')

        # Relative URL
        try:
            relative_url.append(result.find('a', attrs = {'class' : 'offerboxtitle'}).get('href'))
        except:
            relative_url.append('N/A')

        # Product details
        try:
            product_description.append(result.find('div', attrs = {'class' : 'productInfo'}).text.strip().replace('\n', ', '))
        except:
            product_description.append('N/A')
    
```

```python

url_combined = []

for link in relative_url:
    url_combined.append(Root_url + link)

url_combined
```

```python
products_overview = pd.DataFrame({'Product Name' : product_name, 'Product Price' : product_price, 'Review Rating' : review_rating, 'Review Count' : review_count, 'Product Description' : product_description, 'Product URL' : url_combined})
products_overview
```
